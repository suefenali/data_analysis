```python
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.layouts import gridplot
output_notebook()
```

```python
import pandas as pd
cities_data = pd.read_csv("cities_data.csv")
global_data = pd.read_csv("global_data.csv")
```

```python
def calculate_moving_averages(city_df, window_size):
    moving_av = pd.DataFrame(city_df['avg_temp'].rolling(window=window_size).mean())
    moving_av.columns = ['moving_averages']
    return moving_av
```

```python
guarulhos5 = cities_data.loc[cities_data['city'] == 'Guarulhos']
guarulhos15 = cities_data.loc[cities_data['city'] == 'Guarulhos']
guarulhos30 = cities_data.loc[cities_data['city'] == 'Guarulhos']
guarulhos50 = cities_data.loc[cities_data['city'] == 'Guarulhos']

global5 = global_data
global15 = global_data
global30 = global_data
global50 = global_data

guarulhos5 = guarulhos5.join(calculate_moving_averages(guarulhos5, 5))
guarulhos15 = guarulhos15.join(calculate_moving_averages(guarulhos15, 15))
guarulhos30 = guarulhos30.join(calculate_moving_averages(guarulhos30, 30))
guarulhos50 = guarulhos50.join(calculate_moving_averages(guarulhos50, 50))

global5 = global5.join(calculate_moving_averages(global5, 5))
global15 = global15.join(calculate_moving_averages(global15, 15))
global30 = global30.join(calculate_moving_averages(global30, 30))
global50 = global50.join(calculate_moving_averages(global50, 50))
```

```python
def create_figure(guarulhos_df, global_df, title_chart):
    p = figure(plot_width=800, plot_height=600, title=title_chart)
    p.title.text_font_size = '24pt'
    p.line(guarulhos_df['year'], guarulhos_df['moving_averages'], color="navy", alpha=0.8, legend="Guarulhos", line_width=5)
    p.line(global_df['year'], global_df['moving_averages'], color="firebrick", alpha=0.8, legend="Global", line_width=5)
    p.xaxis.axis_label = "Year"
    p.yaxis.axis_label = "Average Temperature"
    p.legend.location = (30,370)
    p.legend.label_text_font_size = '20pt'
    p.xaxis.axis_label_text_font_size = '18pt'
    p.yaxis.axis_label_text_font_size = '18pt'
    p.xaxis.major_label_text_font_size = '18pt'
    p.yaxis.major_label_text_font_size = '18pt'
    p.legend.glyph_width = 50
    p.legend.glyph_height = 40

    return p
```

```python
# Create line chart

p_5_year = create_figure(guarulhos5, global5, "5 years moving average")
p_15_year = create_figure(guarulhos15, global15, "15 years moving average")
p_30_year = create_figure(guarulhos30, global30, "30 years moving average")
p_50_year = create_figure(guarulhos50, global50, "50 years moving average")

grid = gridplot([[p_5_year, p_15_year], [p_30_year, p_50_year]])
show(grid)
```
